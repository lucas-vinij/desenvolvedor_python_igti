# Estrutura de Repeticao 
While e For in

```python
n = 0
while n <= 2022:
  print(f'ERROR {n}')
  n += 1
```

```python
lista = list(range(0,2022 +1))
for n in lista:
  print(n)
```

```python
n = 15
soma = 0
contador = 0

while contador <= n:
  soma += contador
  contador += 1

print(f'A soma dos primeiros {n} inteiros é {soma}') 
```

```python
palavra = 'Araraquara'
count = 0
for i in palavra:
  if i == 'a' or i == 'A':
    count += 1 
print(f'{palavra} tem {count} letras A')
```
